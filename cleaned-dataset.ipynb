# Cleaning dataset

Importing dataset as Pandas Dataframe

```python
df = pd.read_csv(r"/kaggle/input/github-indian-users-deep-data/Github Indian users deep data.csv")
```

### Understanding our data

```python
df.shape
```

```python
df.info()
```

```python
# Null values column wise

df.isnull().sum()
```

#### Dropping column ['Unnamed']

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

### Removing private accounts

```python
df.dropna(subset='last_year_contributions', inplace=True)
```

## Handling Columns

#### **Column:** *followings*

```python
def replace_alphabets_from_values(v: str) -> float :
    """ Return the mathematical values of some values like (1.3k -> 1300) """
    try:
        value = float(v)
    except ValueError:
        value = float(v.replace('k','')) * 1000
    return value
        
```

```python
df.followings = df.followings.apply(replace_alphabets_from_values)
```

#### **Column**: *joining_year*

```python
df.joining_year = df.joining_year.astype('int32')
```

```python
df.info()
```

#### **Column:** *last_year_contributions*

```python
df.last_year_contributions.describe()
```

#### **Column:** *achievement_num*m

```python
df.achievements_num.isnull().sum()
```

#### **Column:** *stars*

```python
df.stars = df.stars.apply(replace_alphabets_from_values).astype('int32')
```

#### **Column:** *repositories*

```python
df.repositories = df.repositories.apply(replace_alphabets_from_values).astype('int')
```

#### **Column:** *followers*

```python
df.followers = df.followers.apply(replace_alphabets_from_values).astype('int32')
```
